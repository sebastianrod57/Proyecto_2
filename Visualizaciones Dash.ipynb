```python
# Consulta para obtener el número de estudiantes becados que se graduaron por género
query_becados_graduados = """
    SELECT Gender, COUNT(*) as conteo_becados_graduados
    FROM Estudiantes
    WHERE Scholarship_holder = 1 AND Target = 1
    GROUP BY Gender;
"""

# Consulta para obtener el número de estudiantes becados que abandonaron por género
query_becados_desertores = """
    SELECT Gender, COUNT(*) as conteo_becados_desertores
    FROM Estudiantes
    WHERE Scholarship_holder = 1 AND Target = 0
    GROUP BY Gender;
"""

# Consulta para obtener el número de estudiantes no becados que se graduaron por género
query_no_becados_graduados = """
    SELECT Gender, COUNT(*) as conteo_no_becados_graduados
    FROM Estudiantes
    WHERE Scholarship_holder = 0 AND Target = 1
    GROUP BY Gender;
"""

# Consulta para obtener el número de estudiantes no becados que desertaron por género
query_no_becados_desertores = """
    SELECT Gender, COUNT(*) as conteo_no_becados_desertores
    FROM Estudiantes
    WHERE Scholarship_holder = 0 AND Target = 0
    GROUP BY Gender;
"""

# Ejecutar las consultas
cursor.execute(query_becados_graduados)
becados_graduados = cursor.fetchall()

cursor.execute(query_becados_desertores)
becados_desertores = cursor.fetchall()

cursor.execute(query_no_becados_graduados)
no_becados_graduados = cursor.fetchall()

cursor.execute(query_no_becados_desertores)
no_becados_desertores =cursor.fetchall()

# Mostrar resultados
print("Número de estudiantes becados que se graduaron por género:")
print(becados_graduados)

print("\nNúmero de estudiantes becados que abandonaron por género:")
print(becados_desertores)

print("\nNúmero de estudiantes no becados que se graduaron por género:")
print(no_becados_graduados)

print("\nNúmero de estudiantes no becados que abandonaron por género:")
print(no_becados_desertores)
# Cerrar el cursor y la conexión con la base de datos
#cursor.close()
#engine.close()
```

```python
# Consulta para obtener el número de estudiantes deudores que se graduaron por género
query_deudores_graduados = """
    SELECT Gender, COUNT(*) as conteo_deudores_graduados
    FROM Estudiantes
    WHERE Debtor = 1 AND Target = 1
    GROUP BY Gender;
"""

# Consulta para obtener el número de estudiantes deudores que abandonaron por género
query_deudores_desertores = """
    SELECT Gender, COUNT(*) as conteo_deudores_desertores
    FROM Estudiantes
    WHERE Debtor = 1 AND Target = 0
    GROUP BY Gender;
"""

# Consulta para obtener el número de estudiantes no deudores que se graduaron por género
query_no_deudores_graduados = """
    SELECT Gender, COUNT(*) as conteo_no_deudores_graduados
    FROM Estudiantes
    WHERE Debtor = 0 AND Target = 1
    GROUP BY Gender;
"""

# Consulta para obtener el número de estudiantes no deudores que abandonaron por género
query_no_deudores_desertores = """
    SELECT Gender, COUNT(*) as conteo_no_deudores_desertores
    FROM Estudiantes
    WHERE Debtor = 0 AND Target = 0
    GROUP BY Gender;
"""

# Ejecutar las consultas
cursor.execute(query_deudores_graduados)
deudores_graduados = cursor.fetchall()

cursor.execute(query_deudores_desertores)
deudores_desertores = cursor.fetchall()

cursor.execute(query_no_deudores_graduados)
no_deudores_graduados = cursor.fetchall()

cursor.execute(query_no_deudores_desertores)
no_deudores_desertores = cursor.fetchall()

# Mostrar resultados
print("Número de estudiantes deudores que se graduaron por género:")
print(deudores_graduados)

print("\nNúmero de estudiantes deudores que abandonaron por género:")
print(deudores_desertores)

print("\nNúmero de estudiantes no deudores que se graduaron por género:")
print(no_deudores_graduados)

print("\nNúmero de estudiantes no deudores que abandonaron por género:")
print(no_deudores_desertores)

# Cerrar el cursor y la conexión con la base de datos
#cursor.close()
#engine.close()
```

```python
import dash
from dash import dcc, html
import plotly.graph_objs as go

# Supongamos que tienes estos datos obtenidos de tus consultas
graduados_por_genero = [(0, 79), (1, 22)]
abandonos_por_genero = [(0, 163), (1, 149)]

# Organizar los datos según el género (0 para mujeres, 1 para hombres)
generos = ['Mujeres', 'Hombres']

graduados_mujeres, graduados_hombres = [item[1] for item in graduados_por_genero if item[0] == 0], [item[1] for item in graduados_por_genero if item[0] == 1]
abandonos_mujeres, abandonos_hombres = [item[1] for item in abandonos_por_genero if item[0] == 0], [item[1] for item in abandonos_por_genero if item[0] == 1]

# Crear la figura con Plotly
fig = go.Figure(data=[
    go.Bar(name='Graduados - Mujeres', x=generos, y=graduados_mujeres, marker=dict(color='red')),
    go.Bar(name='Graduados - Hombres', x=generos, y=graduados_hombres, marker=dict(color='blue')),
    go.Bar(name='Abandonos - Mujeres', x=generos, y=abandonos_mujeres, marker=dict(color='lightcoral')),
    go.Bar(name='Abandonos - Hombres', x=generos, y=abandonos_hombres, marker=dict(color='lightskyblue'))
])

# Actualizar el diseño del gráfico
fig.update_layout(barmode='group', xaxis={'title': 'Género'}, yaxis={'title': 'Cantidad'}, title='Estudiantes Deudores Graduados y Abandonos por Género')

# Iniciar la aplicación Dash
app = dash.Dash(__name__)

# Diseño del tablero
app.layout = html.Div([
    dcc.Graph(figure=fig)  # Agregar el gráfico al layout
])

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```
